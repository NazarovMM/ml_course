```python
# загрузка набора данных
data=pd.read_csv('restaurant-scores-lives-standard.csv', sep=",")
# размер набора данных
data.shape
```

## Обработка пропусков в данных

### 1) обработка пропусков в данных для категориальном признаке

```python
# Выберем категориальные колонки с пропущенными значениями
# Цикл по колонкам датасета
cat_cols = []
for col in data.columns:
    # Количество пустых значений 
    temp_null_count = data[data[col].isnull()].shape[0]
    dt = str(data[col].dtype)
    if temp_null_count>0 and (dt=='object'):
        cat_cols.append(col)
        temp_perc = round((temp_null_count / total_count) * 100.0, 2)
        print('Колонка {}. Тип данных {}. Количество пустых значений {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

```python
cat_temp_data = data[['business_postal_code']]
cat_temp_data.head()
```

```python
cat_temp_data['business_postal_code'].unique()
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Импьютация наиболее частыми значениями
imp2 = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
data_imp2 = imp2.fit_transform(cat_temp_data)
data_imp2
```

```python
# Пустые значения отсутствуют
np.unique(data_imp2)
```

### 2) обработка пропусков в данных для количественного признака

```python
# Выберем числовую колонку с пропущенными значениями
# Цикл по колонкам датасета
num_cols = []
for col in data.columns:
    # Количество пустых значений 
    temp_null_count = data[data[col].isnull()].shape[0]
    dt = str(data[col].dtype)
    if temp_null_count>0 and (dt=='float64' or dt=='int64'):
        num_cols.append(col)
        temp_perc = round((temp_null_count / total_count) * 100.0, 2)
        print('Колонка {}. Тип данных {}. Количество пустых значений {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

Используем различные стратегия для обработки пропусков в колонке inspection_score

Построим график распределения оценок инспектора

```python
sns.distplot(data['inspection_score'])
```

Применим стратегию заполнения 0 всех пропусков

```python
sns.distplot(data['inspection_score'].fillna(0))
```

Пременим стратегию заполнения пропусков среднестатистическим значением

```python
sns.distplot(data["inspection_score"].fillna(data["inspection_score"].mean()))
```

```python
Используя класс SimpleImputer библиотеки sklearn, применим стратегию заполнения пропусков значением медианы
```

```python
mediana = sklearn.impute.SimpleImputer(strategy="median")
median_rating = mediana.fit_transform(data[["inspection_score"]])
sns.distplot(median_rating)
```

```python
data["inspection_score"] = median_rating
data["inspection_score"].isnull().sum()
```
